```python
import sklearn.metrics
from model import model
import numpy

currentModel = model('SCINet_data_ftS_sl48_ll24_pl24_lr0.003_bs8_hid4_s1_l3_dp0.5_invFalse_itr0')
dateStart = '2020-10-24 11:00:00'
dateEnd = '2020-10-25 11:00:00'
pred, true, pred_scale, true_scale, dates = currentModel.predict(dateStart, dateEnd)
preds = []
trues = []
pred_scales = []
true_scales = []

for item in pred:
    preds.append(item.mean())
for item in true:
    trues.append(item.mean())
for item in pred_scale:
    pred_scales.append(item.mean())
for item in true_scale:
    true_scales.append(item.mean())
```

```python
def get_classifiers(threshold):
  TP = 0
  FP = 0
  TN = 0
  FN = 0
  for i in range(len(pred_scales)):
    if pred_scales[i] >= threshold:
      if true_scales[i] >= threshold:
        TP = TP + 1
      else:
        FP = FP + 1
    else:
      if true_scales[i] < threshold:
        TN = TN + 1
      else:
        FN = FN + 1
  return (TP, FP, TN, FN)
```

```python
def get_threshold(percent):
    result = 0
    threshold = REGALATORY_THRESHOLD
    while result < percent:
        TP, FP, TN, FN = get_classifiers(threshold)
        result = ((TP + FN) / len(pred_scales)) * 100
        threshold = threshold - 1
    return threshold    
```

```python
REGALATORY_THRESHOLD = 75
PERCENT_OF_DEFECTS = 40
```

```python
TP, FP, TN, FN = get_classifiers(get_threshold(PERCENT_OF_DEFECTS))
TPR = TP / (TP + FN)
FPR = FP / (FP + TN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
F = 2 * ((precision * recall) / (precision + recall))
```

```python
print('precision = {0}, recall = {1}, F1 = {2}'.format(precision, recall, F))
```

```python
TPR = []
FPR = []

thresholds = numpy.linspace(0, REGALATORY_THRESHOLD)

for i in thresholds:
  tp, fp, tn, fn = get_classifiers(i)
  if (tp + fn) != 0:
    TPR.append(tp / (tp + fn))
  else:
    TPR.append(1)
  if (fp + tn) != 0:
    FPR.append(fp / (fp + tn))
  else:
    FPR.append(1)

TPR.sort()
FPR.sort()

AUC = 0
for i in range(len(thresholds) - 1):
  AUC = AUC + (((TPR[i] + TPR[i + 1]) / 2) * (FPR[i + 1] - FPR[i]))
```
